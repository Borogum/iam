Para medir el progreso de las distintas tareas vamos a definir la siguiente clase:

````python
import math

class ProgressBar(object):
    
    def __init__(self, count, steps, task='Task', marker='=='):

        self.end = count
        self.steps = steps
        self.last_step = 0
        self.current_step = 0
        self.current = 0
        self.step_size = count / steps
        self.marker = marker
        self.showed = False
        self.task = task
    
    def __iadd__(self, other):
        
        self.current = min(self.current + other, self.end)
        self.current_step = math.floor(self.current / self.step_size)
        return self
    
    def show(self):
        
        if self.last_step < self.steps:
            if not self.showed:
                print('{}: |'.format(self.task), end='')
                self.showed = True
            print(self.marker * (self.current_step - self.last_step) , end='')
            self.last_step = self.current_step
            if self.current_step == self.steps:
                print('| Done!')
````

Su funci√≥n es imprimir por pantalla una barra de progreso para tener un _feedback_ en aquellas tareas que tardan en ejecutarse (como la descarga de las im√°genes). Solo tiene un fin est√©tico por lo que no es importante para la compresi√≥n del resto del proceso.

Despu√©s de un exhaustivo an√°lisis del c√≥digo _HTML_, he averiguado que para poder descargarnos las im√°genes de los tickets necesitamos dos identificadores:

- Un identificador √∫nico que llamaremos _token_. Este se obtiene de los enlaces a los tickets (üëÅÔ∏è).

- Otro identificador que no s√© muy bien como catalogar que se denomina _plid_. Para obtener este valor es necesario extraerlo directamente del c√≥digo _HTML_. En particular, se encuentra en un fragmento de c√≥digo JavaScript.

El _helper_ que definimos a continuaci√≥n, sirve para obtener el segundo de los identificadores.

````python
def get_plid(bs):
    
    """Extract plid from JavaScript"""
    
    id_regex = r'getPlid:\s*function\(\)\s*\{\s*return\s+"(\d+)";\s*}'
    for scr in bs.find_all('script'):
        if scr.string is not None:
            m = re.search(id_regex, scr.string)
            if m:
                plid = m.group(1)
                return plid
    return None
````

Como ya mencionamos anteriormente el uso de la _API_ de _Google_ es muy sencillo. A continuaci√≥n, pod√©is ver la funci√≥n que se encarga de extraer el texto de cada imagen ¬°solamente seis l√≠neas!:

````python
def detect_text(client, image_path):
    """Detects text in the file."""
    with io.open(image_path, 'rb') as image_file:  # Open image file
        content = image_file.read()  # Read image content
    response = client.document_text_detection(image=vision.Image(content=content))  # Call API
    return response.text_annotations  # Return annotations
````

Lo √∫nico que necesitamos ahora es usar funci√≥n ```detect_text``` en cada una de las im√°genes descargadas y almacenar el resultado para su posterior procesado. El fragmento de c√≥digo responsable de esta tarea ser√≠a el siguiente:

````python
print('Creating API client: ', end='')
client = vision.ImageAnnotatorClient.from_service_account_json(VISION_AUTH_FILE) # Create client from JSON auth file.
print('Done!')

progress = ProgressBar(len(tickets_summary_df), 10, task='Processing tickets')

for filename in glob.glob(os.path.join(TICKETS_OUTPUT_DIR, '*.jpg')):
    ticket_id = os.path.basename(filename).split('.')[0]
    texts = detect_text(client, filename)
    data = []
    for text in texts:
        vertices = ([(vertex.x, vertex.y) for vertex in text.bounding_poly.vertices])
        data.append({'text': text.description, 'vertices': vertices})  #  Only relevant info is stored
    # Save data into JSON format
    with open(os.path.join(VISION_OUTPUT_DIR, os.path.basename(filename).split('.')[0] + '.json'), 'w') as outfile:
        json.dump(data, outfile)
    progress += 1
    progress.show()
````

Hablando en t√©rminos de estructuras de datos de _Python_ b√°sicamente es un [lista](https://www.w3schools.com/python/python_lists.asp) de [diccionarios](https://www.w3schools.com/python/python_dictionaries.asp), donde cada diccionario contiene dos entradas:

- **text:** el texto.
- **vertices:** los v√©rtices de la _bounding box_ (cajita que contiene al texto üì¶) en el siguiente orden: sup-izq, sup-der, inf-der y inf-izq.

Ahora disponemos de dos formas de representar la misma informaci√≥n: una en forma de imagen (üëé) y otra en forma de palabras y cajitas (üëç). Para se vea m√°s claramente las diferencias/similitudes entre las dos, vamos a representar ambas para compararlas. Definiremos un par de funciones que nos ayudar√°n a representar gr√°ficamente todas esas palabras. La primera, ```plot_word```, plotea cada palabra con su _bounding box_. La segunda, ```plot_ticket```, √∫nicamente aplica la funci√≥n anterior a todas las palabras extra√≠das de un determinado ticket.

````python
from matplotlib.patches import Rectangle


def plot_word(word, color, ax, size=8, axhline=False):
    
    """Plots text with its bounding box"""
    
    ax.text(*word['vertices'][0], word['text'], va='top', ha='left', size=size)  # Plot text
    w = word['vertices'][1][0] - word['vertices'][0][0]
    h = word['vertices'][-1][1] - word['vertices'][0][1]
    rect = Rectangle(word['vertices'][0], w, h, linewidth=1, facecolor=color, alpha=0.3)  # Create bounding box
    if axhline:
        ax.axhline(y=word['vertices'][0][1] + h/2, color=color, alpha=0.6)  # Plot horizotal line
    ax.add_patch(rect)  # Plot box


def plot_ticket(ticket_words, width, height, ax):
    
    """Plots all words of one ticket"""
    
    colors = ['blue', 'red', 'green', 'yellow', 'grey', 'orange']

    for word in ticket_words:
        color = np.random.choice(colors, 1)[0]
        plot_word(word, color, ax)
    
    ax.set(xlim=(0, width), ylim=(0, height), xticks=(), yticks=())
    ax.invert_yaxis()
````

<a id="structuring"></a>

#### Estructuraci√≥n del texto

En este punto tenemos un mont√≥n de palabras sueltas desperdigadas por el espacio (textos y cajitas). Un ticket, como cualquier documento escrito, tiene l√≥gica cuando sus palabras est√°n organizadas de forma coherente y no flotando por ah√≠. La forma de organizar estas palabras son las l√≠neas. Es f√°cil leer una l√≠nea, solo tenemos que recorrer las palabras en un orden espec√≠fico (de arriba a abajo y de izquierda a derecha en el caso del castellano). El objetivo de este apartado es precisamente ese, agrupar esas "palabras flotantes" en l√≠neas.

Como conocemos las posiciones de las cajas que contienen los textos (de ah√≠ la importancia de guardar esos valores) podemos establecer un criterio muy sencillo pero efectivo para saber si dos palabras pertenecen a la misma l√≠nea que es el siguiente: dos palabras pertenecen a la misma l√≠nea si la distancia vertical entre sus puntos medios es menor que cierto valor prefijado al que llamaremos umbral. Vamos a ilustrar el concepto con un ejemplo muy sencillo:

````python
def make_vertices(x, y, width, height):
    """Create the vertices of a bounding box with: x, y, width and height values"""
    return [[x, y], [x + width, y], [x + width, y - height], [x, y - height]]


letter_width = 5.6
letter_height = 1
colors = ['blue', 'red', 'green', 'yellow', 'gray']

# Some fake data simulating JSON files
text_data = [{'text': 'Hola', 'vertices': make_vertices(0 , 4, letter_width * 4, letter_height)},
        {'text': 'a', 'vertices': make_vertices(27 , 4.2, letter_width * 1, letter_height)},
        {'text': 'todo', 'vertices': make_vertices(40 , 3.9, letter_width * 4, letter_height)},
        {'text': 'el', 'vertices': make_vertices(65 , 3, letter_width * 2, letter_height)},
        {'text': 'mundo', 'vertices': make_vertices(80 , 4.1, letter_width * 7, letter_height)},
       ]

fig, ax = plt.subplots(figsize=(20, 5))
ax.set(xlim=(-5, 124), ylim=(1, 5), xticks=(), yticks=()) 

for color, word in zip(colors, text_data):
    plot_word(word, color, ax, size=80, axhline=True)
````

En este ejemplo las palabras: _"Hola"_, _"a"_, _"todo"_, _"mundo"_ pertenecer√≠an a una misma l√≠nea (sus centros est√°n muy pr√≥ximos) mientras que la palabra _"el"_ pertenecer√≠a a una l√≠nea distinta. Entendido el concepto, ser√° muy f√°cil comprender el proceso que transformar√° nuestra nube de palabras en un conjunto de l√≠neas.

Antes de pasar a ver el c√≥digo, vamos a estudiar la importancia de elegir un buen _threshold_ (umbral). Usaremos una funci√≥n _helper_ para mostrar visualmente los resultados. Esta nueva funci√≥n est√° basada en ```plot_ticket``` (ver apartado anterior) modificada levemente para mostrar las l√≠neas obtenidas (l√≠neas horizontales grises) en funci√≥n del _threshold_.

````python
def plot_ticket_with_lines(ticket_words, threshold, width, height, ax):
    
    """Plots all words of one ticket with lines"""
    
    colors = ['blue', 'red', 'green', 'yellow', 'grey', 'orange']
    last_line = 0
    ticket_words = sorted(ticket_words, key=lambda x: x['vertices'][0][1], reverse=False)  # From top to bottom (y-axis) 
    
    for word in ticket_words:
        color = np.random.choice(colors, 1)[0]
        plot_word(word, color, ax)
        this_line = (word['vertices'][-1][1] + word['vertices'][0][1])/2
        if (this_line - last_line) > threshold: 
            ax.axhline(y=this_line, linewidth=0.5, color='black', alpha=0.7)
            last_line = this_line

    ax.set(xlim=(0, width), ylim=(0, height), xticks=(), yticks=())
    ax.invert_yaxis()
````

Veamos ahora como var√≠an los resultados al usar distintos _thresholds_:

Definiremos dos funciones auxiliares para descomponer las tareas y hacer el c√≥digo m√°s legible. La primera, ```parse_number```, es muy simple y transforma las cadenas de texto que representan n√∫meros en n√∫meros de verdad. Por ejemplo, las cadenas: _"2"_, _"1,8"_, _"1, 00"_ o _"1 , 2"_ se transformar√≠an respectivamente en: *2.0*,  *1.8*, *1.0* y *1.2*.

````python
def parse_number(number):
    """Parses  'numbers' to real numbers"""
    return float(str(number).replace(',', '.').replace(' ', ''))
````

La segunda, ```parse_ticket```, se encarga de transformar los tickets en formato texto (lista de cadenas de texto) en un listado de diccionarios. Donde cada diccionario almacenar√° la informaci√≥n de la venta de cada producto:

* __ticket_id:__ identificador √∫nico del ticket.
* __product_name:__ nombre del producto.
* __iva:__ tipo de iva aplicado.
* __units:__ unidad de medida del producto KG/UNIDADES.
* __count:__ n√∫mero de unidades o peso. 
* __unit_price:__ precio por unidad o por KG.
* __total_price:__ coste total del producto. Si todo es correcto deber√≠a ser igual a  __count__ * __unit_price__.
* __discount_name:__ nombre o nombres de los descuentos aplicados separados por comas. 
* __discount:__ importe del descuento.
* __total_ticket:__ suma total del ticket al que pertenece la venta.

A continuaci√≥n, se muestra el c√≥digo de la funci√≥n. Esta ampliamente comentado para aquellos que tengan curiosidad por los detalles de la implementaci√≥n:

````python
def parse_ticket(ticket, ticket_id):
    
    """Transforms a ticket (list of lines) into a list of dictations with the information of the purchases"""

    # Regular expressions (re). Same name conventions as exposed in examples
    simple_line = r'^(?P<product_name>.*)\s+(?P<unit_price>\d+\s*,\s*\d{2})\s?(?P<iva>[0-9A-Z])?$'
    count_line = r'^(?P<count>\d+(\s*,\s*\d+)?)\s+(KG\s+)?X\s+(?P<unit_price>\d+\s*,\s*\d{2})\s+EUR \/ ' + \
                 r'(?P<units>(UNIDAD|KG))\s+(?P<total_price>\d+\s*,\s*\d{2})\s?(?P<iva>[0-9A-Z])?$'
    discount_line = r'^(?P<discount_name>.*)\s+-(?P<discount>\d+\s*,\s*\d{2})$'
    total_line = r'^TOTAL\s+[^0-9]*(?P<total>\d+\s*,\s*\d{2})\s+EUR$'

    # Compiled regular expressions are stored in a dictionary for convenience
    re_dict = {'simple_line': re.compile(simple_line),
               'count_line': re.compile(count_line),
               'discount_line': re.compile(discount_line),
               'total_line': re.compile(total_line)}

    # In all tickets the top four lines contains supermarket info. No relevant for us
    skip = 4  # Skip loc info

    purchases = []  # List for all purchases 
    purchase = {}  # One purchase

    for ticket_line in ticket[skip:]:  # For each line (Skip first 4 lines)

        current_line = ticket_line.upper().strip()  # String cleaning
        # All the r.e. are applied to line to find out the line type. 
        searches = {k: re_dict[k].search(current_line) for k in re_dict}
        # The order of evaluation is relevant!
        if searches['count_line']:  # Is a count line. Update purchase info.
            purchase.update(searches['count_line'].groupdict())

        elif searches['discount_line']:  # Is a discount line. There are two possibilities now:

            if 'discount' in purchase:  # Previous discount exists.
                # Acceptable solution only if u are not interested in individual discounts.
                # TODO implement a better approach
                new_discount = searches['discount_line'].groupdict()  # Get discount info
                # New discount string is added to existing discount string separated by a comma
                purchase['discount_name'] += ', ' + new_discount['discount_name']
                # Increment discount amount
                purchase['discount'] = parse_number(purchase['discount']) + parse_number(
                    new_discount['discount'])  # No problem if discount is already a number
            else:  # There is not previous discount.
                purchase.update(searches['discount_line'].groupdict())  # Update purchase info

        elif searches['simple_line']:  # Is a simple line so all previous stored info if exists, refers to another purchase
            if len(purchase):  # If there is a info from the previous purchase
                purchases.append(purchase)  # Save it in purchases list
            # Create a new purchase
            purchase = searches['simple_line'].groupdict()
            purchase['count'] = 1
            purchase['total_price'] = purchase['unit_price']
            purchase['units'] = 'UNIDAD'

        elif searches['total_line']:  # Is a total line. 
            # Store total ticket cost and break the loop. Almost all done.
            total_ticket = searches['total_line'].group('total')
            break

        else:  # If is not a "another thing" is a multiple line. All previous stored info if exists, refers to another purcha
            if len(purchase):  # If there is a info from the previous purchase
                purchases.append(purchase)  # Save it in purchases list
            # Create a new purchase
            purchase = {'product_name': current_line}

    purchases.append(purchase)  # Store last purchase created
    # Add to all purchases total cost and ticket id info
    for purchase in purchases:
        purchase['total_ticket'] = total_ticket
        purchase['ticket_id'] = ticket_id

        # Parse 'number' attribute to number attribute
        for att in ['unit_price', 'count', 'total_price', 'discount', 'total_ticket']:
            if att in purchase:  # Note that discount is not present in all records
                purchase[att] = parse_number(purchase[att])

    return purchases
````

````python
from dateutil import relativedelta

start_date = plastic_bags_df.ticket_date.min()
end_date = plastic_bags_df.ticket_date.max()
months = relativedelta.relativedelta(end_date, start_date).months
mrkdwn = ''

fact= '- He utilizado un total de __{total}__ bolsas de pl√°stico en un periodo de aproximadamente __{months}__ meses.\r\n'
mrkdwn += fact.format(total=plastic_bags_df['count'].sum().astype(int), months=months)

fact= '- La estimaci√≥n de consumo ser√≠a de __{est:d}__ bolsas de pl√°stico por a√±o.\r\n'
mrkdwn += fact.format(est=int(round(plastic_bags_df['count'].sum() * 12 /8)))

fact= '- El promedio de bolsas compradas (cuando las he "necesitado") ha sido de __{bag_mean:.2f}__ por visita.\r\n'
mrkdwn += fact.format(bag_mean=plastic_bags_df['count'].sum()/plastic_bags_df.ticket_id.nunique())

fact= '- He gastado __{bag_cost:.2f} ‚Ç¨__ en bolsas.\r\n'
mrkdwn += fact.format(bag_cost=plastic_bags_df.total_price.sum())

fact= '- Desde el __{date}__ no he comprado ninguna bolsa de pl√°stico m√°s.\r\n'
mrkdwn += fact.format(date=end_date.strftime('%d/%m/%Y'))

md(mrkdwn)
````
